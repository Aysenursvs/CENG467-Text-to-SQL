```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import os

# Spider veri setini yükle
dataset = load_dataset("xlangai/spider")
train_df = pd.DataFrame(dataset['train'])
val_df = pd.DataFrame(dataset['validation'])

print(f"Eğitim Seti Boyutu: {len(train_df)}")
print(f"Doğrulama Seti Boyutu: {len(val_df)}")
```

```python
# Soruların kelime sayılarını hesapla
train_df['question_length'] = train_df['question'].apply(lambda x: len(x.split()))
train_df['sql_length'] = train_df['query'].apply(lambda x: len(x.split()))

print("--- Rapor İçin İstatistikler ---")
print(f"Ortalama Soru Uzunluğu: {train_df['question_length'].mean():.2f} kelime")
print(f"Maksimum Soru Uzunluğu: {train_df['question_length'].max()} kelime")
print(f"Ortalama SQL Uzunluğu: {train_df['sql_length'].mean():.2f} token/kelime")
print(f"Maksimum SQL Uzunluğu: {train_df['sql_length'].max()} token/kelime")

# Maksimum uzunluktaki SQL'in ne kadar karmaşık olduğuna bakalım:
longest_sql = train_df.loc[train_df['sql_length'].idxmax()]
print(f"\nEn uzun SQL Örneği (DB: {longest_sql['db_id']}):\n{longest_sql['query']}")
```

```python
train_dbs = set(train_df['db_id'].unique())
val_dbs = set(val_df['db_id'].unique())

common_dbs = train_dbs.intersection(val_dbs)

print(f"Eğitim setindeki eşsiz veritabanı sayısı: {len(train_dbs)}")
print(f"Doğrulama setindeki eşsiz veritabanı sayısı: {len(val_dbs)}")
print(f"Hem eğitimde hem doğrulamada olan ortak veritabanı sayısı: {len(common_dbs)}")
# Eğer sonuç 0 çıkarsa, bu "Cross-Domain" zorluğunu mükemmel bir şekilde kanıtlar!
```
